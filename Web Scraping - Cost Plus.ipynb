This ipynb file contains the code used to (legally) webscrape the Cost Plug Drugs website for the price data contained in the costplus csv.

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://costplusdrugs.com/medications/")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
```

```python
medication_name_class = "items-center flex-shrink-0 p-md border-2 tablet:border-none border-cap-gray-100 w-6xl tablet:w-1/3 tablet:flex-row bg-cap-white sticky left-0 flex flex-col text-left"
dosage_form_class = "items-center flex-shrink-0 p-md border-2 tablet:border-none border-cap-gray-100 w-6xl tablet:w-12% tablet:flex justify-center text-center pt-2xl tablet:pt-md"
```

```python
#Scraping medication names
medication = []
for element in soup.findAll(attrs={'class': medication_name_class}):
    name = element.find('a')
    medication.append(name.text)
medication
```

```python
len(medication)
```

```python
#Scraping dosage form, retail price, CP price (all three are under the same class, occuring at certain index multiples
form = []
for element in soup.findAll(attrs={'class': dosage_form_class}):
    form.append(element.text)
dosage_form = form[::4]
dosage_form;
```

```python
len(dosage_form)
```

```python
#Retail Price
retail_price = form[1::4]
retail_price;
```

```python
len(retail_price)
```

```python
#CP price
CP_price = form[2::4]
CP_price;
```

```python
len(CP_price)
```

```python
costplus = pd.DataFrame({"medication": medication, "form": dosage_form, "retail_price": retail_price, "CP_price": CP_price})
costplus.head()
```

```python
#Converting retail price and CP price to floats
costplus["retail_price"] = costplus["retail_price"].str.split('$').str[1].str.replace(',', '').astype(float)
costplus["CP_price"] = costplus["CP_price"].str.split('$').str[1].str.replace(',', '').astype(float)
costplus.head()
```

```python
type(costplus["retail_price"][0])
type(costplus["CP_price"][0])
```

```python
#Create a column which contains brand name equivalent
costplus["brand_name"] = costplus["medication"].str.extract(r"\(Generic\sfor\s([A-Z]{1}[a-z]*)\)")
```

```python
costplus = costplus[["medication", "brand_name", "form", "retail_price", "CP_price"]]
costplus.head()
```
